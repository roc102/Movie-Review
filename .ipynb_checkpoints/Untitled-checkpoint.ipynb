```python
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

# Load the labeled training dataset (replace 'label.tsv' with the actual file path)
train_data = pd.read_csv('C:/Users/LENOVO/Desktop/Movie-Review/labeledTrainData.tsv', sep='\t')

# Clean the movie reviews using regex
def clean_text(text):
    text = re.sub(r'<br />', ' ', text)  # Remove HTML tags
    text = re.sub(r'\W+', ' ', text)  # Remove non-word characters
    text = text.lower()  # Convert to lowercase
    return text

train_data['review'] = train_data['review'].apply(clean_text)

# Separate features (movie reviews) and target (sentiment: positive or negative)
X_train = train_data['review']
y_train = train_data['sentiment']

# Convert movie reviews into numerical features using CountVectorizer
vectorizer = CountVectorizer(max_features=10000)
X_train = vectorizer.fit_transform(X_train)

# Create and train the Naive Bayes classifier
classifier = MultinomialNB()
classifier.fit(X_train, y_train)

# Load the test dataset (replace 'data.tsv' with the actual file path)
test_data = pd.read_csv('C:/Users/LENOVO/Desktop/Movie-Review/testData.tsv', sep='\t')

# Clean the movie reviews in the test dataset using regex
test_data['review'] = test_data['review'].apply(clean_text)

# Preprocess the text data (if needed)
X_test = test_data['review']

# Convert movie reviews into numerical features using the same CountVectorizer
X_test = vectorizer.transform(X_test)

# Make predictions on the test set
y_pred = classifier.predict(X_test)

# Print the movie reviews and their predicted sentiment
for review, sentiment in zip(test_data['review'], y_pred):
    if sentiment == 1:
        print(f"Review: \nSentiment: Positive\n")
    else:
        print(f"Review: \nSentiment: Negative\n")

# Visualization of positive and negative reviews
sentiment_counts = train_data['sentiment'].value_counts()
plt.bar(sentiment_counts.index, sentiment_counts.values, color=['blue', 'red'])
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.title('Distribution of Positive and Negative Reviews')
plt.xticks([0, 1], ['Negative', 'Positive'])
plt.show()
```
